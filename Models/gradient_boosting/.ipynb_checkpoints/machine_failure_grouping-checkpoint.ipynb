# Machine Failure Grouping
* This notebook separates the training data by whether the machine failure is true (1) or false (0), and saves these as csv files for use in plotting (with Tableau in the case of this project)

```python
import numpy as np
import pandas as pd
```

## Split the Data into Training and Testing Sets

```python
# Load the test data into a Pandas DataFrame
df_train = pd.read_csv("https://machine-failure-data-20230822-craiguo.s3.us-west-2.amazonaws.com/train.csv")
df_train.drop(columns=['id', 'Product ID'], inplace=True)

# Display sample data
df_train.head(10)
```

```python
# Separate by value of 'Machine failure'
df_machine_failure_0 = df_train[df_train['Machine failure']==0]
df_machine_failure_1 = df_train[df_train['Machine failure']==1]
```

```python
# Review the dataframe for machine failure false
df_machine_failure_0.head()
```

```python
# Review the dataframe for machine failure true
df_machine_failure_1.head()
```

```python
# Saving machine failure false to csv
df_machine_failure_0.to_csv("machine_failure_false.csv", index=False)

# Saving machine failure true to csv
df_machine_failure_1.to_csv("machine_failure_true.csv", index=False)
```
